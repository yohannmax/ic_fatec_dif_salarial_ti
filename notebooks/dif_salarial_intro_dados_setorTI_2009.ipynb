### Imports e leitura dos dados:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Opções gerais necessárias:

pd.options.display.float_format = "{:,.2f}".format 
```

```python
# Leitura da tabela principal:
RAIS_2019 = pd.read_csv("../data/raw/DATA_RAIS_2009.csv", sep = ";", encoding = 'utf-8', skipinitialspace = True)

# Renomeando as colunas:
RAIS_2019.rename(columns = {'CBO Ocupação 2002':'CBO', 'CNAE 2.0 Classe':'CNAE',
                            'Escolaridade após 2005':'Escolaridade',
                            'Raça Cor':'Raça', 'Vl Remun Dezembro Nom':'Remuneração',
                            'Sexo Trabalhador':'Sexo'}, inplace = True)

# Exibindo os cinco primeiros dados da tabela:
RAIS_2019.head()
```

### Demonstração dos dados por UF:

```python
dist_freq_UF = pd.DataFrame(
    {'Frequência UF (Masculino)': RAIS_2019.query("Sexo == 1").UF.value_counts(), 
     'Porcentagem (%) (Masculino)': RAIS_2019.query("Sexo == 1").UF.value_counts(normalize = True) * 100,
     'Frequência UF (Feminino)': RAIS_2019.query("Sexo == 2").UF.value_counts(), 
     'Porcentagem (%) (Feminino)': RAIS_2019.query("Sexo == 2").UF.value_counts(normalize = True) * 100,
     }
)

dist_freq_UF.sort_values('Frequência UF (Masculino)', ascending = False).head(3)
```

```python
dist_freq_UF.sort_values('Frequência UF (Masculino)', ascending = True).head(6)
```

Os três maiores estados do país com mais profissionais (tanto masculinos quanto femininos): 
São Paulo com quase 50% de todos os profissionais de TI nos setores de TI do país.
Seguido de Rio de Janeiro, com 11% e 12,5% de profissionais, masculinos e femininos respectivamente, no total dos estados e 
por fim Minas Gerais, com aproximadamente 7% (nos dois sexos).

Quando observamos os estados com menores quantidades de funcionários de TI temos Roraima, Amapá, Acre, Rondonia, Tocantins e 
Piauí com menos de 0,1% do total dos estados, com baixíssima representação de profissionais da área.
